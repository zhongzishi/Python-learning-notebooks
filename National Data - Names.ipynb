```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
names2015 = pd.read_csv(r'C:\Users\Zishi\Downloads\names\yob2015.txt', names=['name','gender','births'])
```

```python
names2015.groupby('gender').births.sum()
```

```python
# concat all files from 1880 - 2015
years = range(1880,2016)
temp = []
cols = ['name','gender','births']
for year in years:
    path = r'C:\Users\Zishi\Downloads\names\yob%d.txt' % year
    frame = pd.read_csv(path,names=cols)
    frame['year'] = year
    temp.append(frame)
names = pd.concat(temp,ignore_index=True)
```

```python
total_births = names.pivot_table('births',index='year',columns='gender',aggfunc=sum)
total_births.plot(title='Total births by gender and year')
```

```python
total_births['diff'] = total_births['F'] - total_births['M']
total_births['diff'].plot(title='Female minus Male');plt.axhline(0,color='r')
```
