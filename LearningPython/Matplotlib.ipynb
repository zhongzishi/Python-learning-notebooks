```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt

a = np.linspace(0,2*np.pi,50)
b = np.sin(a)
plt.plot(a,b,'ro')
```

```python
plt.scatter(a,b,5,'r')
```

```python
x = np.random.rand(200)
y = np.random.rand(200)
size = np.random.rand(200)*30
color = np.random.rand(200)
plt.scatter(x,y,size,color)
plt.colorbar()
```

```python
t = np.linspace(0,2*np.pi,50)
x = np.sin(t)
y = np.cos(t)
plt.figure()
plt.plot(x)
plt.figure()
plt.plot(y)
```

```python
plt.subplot(1,2,1)
plt.plot(x)
plt.subplot(1,2,2)
plt.plot(y)
```

```python
plt.subplot(2,2,1)
plt.plot(x)
plt.subplot(2,2,2)
plt.plot(y)
```

```python
plt.plot(x)
plt.plot(y)
```

```python
plt.plot(x)
plt.hold(False)
plt.plot(y)
plt.hold(True)
```

```python
plt.plot(x,label='sin')
plt.plot(y,label='cos')
plt.legend()
```

```python
plt.plot(x)
plt.plot(y)
plt.legend(['sin','cos'])
```

```python
plt.plot(x,np.sin(x))
plt.xlabel('radians')
plt.ylabel('amplitude',fontsize='large')
plt.title('Sin(x)')
plt.grid()
```
