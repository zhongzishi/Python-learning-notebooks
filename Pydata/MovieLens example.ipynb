```python
import pandas as pd
```

```python
unames = ['user_id','age','gender','occupation','zip']
users = pd.read_table(r'C:\Users\Zishi\Python learning notebooks\Pydata\ml-100k\u.user', sep='|', header=None, names=unames)
```

```python
rnames = ['user_id','movie_id','rating','timestamp']
ratings = pd.read_table(r'C:\Users\Zishi\Python learning notebooks\Pydata\ml-100k\u.data', sep='\t', header=None, names=rnames)
```

```python
ratings.head()
```

```python
mnames = ['movie_id','movietitle','release date','videoreleasedate','IMDbURL']
movies = pd.read_table(r'C:\Users\Zishi\Python learning notebooks\Pydata\ml-100k\u.item',sep='|',names=mnames,
                      usecols = range(5),encoding='ISO-8859-1')
```

```python
movies[:5]
```

```python
data = pd.merge(pd.merge(ratings,users),movies)
```

```python
data[:5]
```

```python
data.ix[0]
```

```python
mean_ratings = data.pivot_table('rating',index='movietitle',
                               columns='gender',aggfunc='mean')
```

```python
mean_ratings[:5]
```

```python
ratings_by_title = data.groupby('movietitle').size()
```

```python
ratings_by_title.head()
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 100]
```

```python
active_titles
```

```python
mean_ratings = mean_ratings.ix[active_titles]
```

```python
mean_ratings
```

```python
top_female_ratings = mean_ratings.sort_values(by='F',ascending=False)
```

```python
top_female_ratings[:30]
```

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
```

```python
sorted_by_diff = mean_ratings.sort_values(by='diff')
```

```python
sorted_by_diff.head()
```

```python
sorted_by_diff[::-1][:15]
```

```python
rating_std_by_title = data.groupby('movietitle')['rating'].std()
```

```python
rating_std_by_title = rating_std_by_title.ix[active_titles]
```

```python
rating_std_by_title.sort_values(ascending=False)[:30]
```
